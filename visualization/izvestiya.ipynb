```julia
include("../src/cracks.jl")
using Plots; gr();
```

```julia
geotherms = [9, 15, 26]
depths = collect(1:40)
means = Dict()
pl = plot()
for g in geotherms
    p = [izvestiya_porosity(g, d) for d in depths]
    println("Mean porosity $(mean(p)) for geotherm $g degC/km")
    means[g] = mean(p)
    plot!(pl, log.(p), depths, label=g)
end 
plot!(yflip=true)
```

```julia
means[26] / means[15]
```
