## Find dry modulus for cracks with aspect ratio a and porosity p 

David and Zimmerman give forms for bulk modulus K and Poisson's ratio nu as functions of the aspect ratio alpha and bulk and shear moduli of the solid K0 and G0 (which can calculate nu0)

```julia
function gamma(porosity, alpha)
    return porosity / (4/3 * pi * alpha)
end

println(poissons(.25,.25))
println(gamma(.1, .2))
```

```julia
## Oblate spheroids 

# alpha = .2 # This form is good for 0 <= alpha <= .3
# nu0 = .25 # Poisson's ratio of solid 
# porosity = .1

c1 = 25/864 * (48/pi + pi)
c2 = 5/288 * (48/pi + 5*pi)
c3 = 5/81 * (82/pi^2 - 163/16 + (1753/1536)*pi^2)
c4 = 5/27 * (8/pi^2 - 61/6 + (137/768)*pi^2)
c5 = 5/9 * (1/pi + 5*pi/48)^2

println(c1)
println(c2)
println(c3)
println(c4)
println(c5)


# Eq 35
function build_poissons_expr(alpha, nu0, porosity)
    return function poissons_expr(nu)
        left = (-3/(4*pi*alpha)) * log(1-(4/3)*pi*alpha*gamma(porosity, alpha))
        right1 = -1*(5/8)*log(nu/nu0)
        right2 = alpha * ( c1*log(nu/nu0) + c2*(1/nu - 1/nu0) )
        right3 = alpha^2 * (c3*log(nu/nu0) + c4*(1/nu - 1/nu0) + c5*(1/nu^2 - 1/nu0^2))
        return (right1 + right2 + right3) - left
    end 
end
```

```julia
myfunc = build_poissons_expr(.2, .25, .1)
find_zero(myfunc, (0.01,.5))
```

```julia
a = .2 
nu0 = .25

nus = []
for i in 0:.01:.8
    print("$i.")
    push!(nus, find_zero(build_poissons_expr(a, nu0, i), (0.01,.5)))
end 
plot([gamma(i,a) for i in 0:.01:.8], nus, label="alpha=$a")

a2 = .01

nus2 = []
for i in 0:.001:.05
    print("$i.")
    push!(nus2, find_zero(build_poissons_expr(a2, nu0, i), (0.01,.5)))
end 

plot!([gamma(i,a2) for i in 0:.001:.05], nus2, label="alpha=$a2")

    
```

## Implement limiting cases for cracks, needles, and spheres

More flexible solutions provide little benefit in our application -- there's not enough data resolution to tell the difference between a crack-like pore with α = .2 vs α = 0.01



```julia
nu0 = .25
nus = []
bulks = [] # K/K0
for porosity in 0:.001:.05
    props = cracked_properties(porosity, nu0)
    push!(nus, props[1])
    push!(bulks, props[2])
end 

## To compare, plot gamma
#p1 = plot([gamma(i,.01) for i in 0:.001:.05], nus, ylabel="nu", xlabel="Crack density gamma", legend=false)
#p2 = plot([gamma(i,.01) for i in 0:.001:.05], bulks, ylabel="K/K0", xlabel="Crack density gamma", legend=false)

# else, plot porosity
p1 = plot(0:.001:.05, nus, ylabel="nu", xlabel="Porosity", legend=false)
p2 = plot(0:.001:.05, bulks, ylabel="K/K0", xlabel="Porosity", legend=false)
plot(p1, p2)
```

```julia
nu0 = .25
nus = []
bulks = [] # K/K0
for porosity in 0:.1:1
    props = needle_properties(porosity, nu0)
    push!(nus, props[1])
    push!(bulks, props[2])
end 

p1 = plot(title = "Needles", 0:.1:1, nus, ylabel="nu", xlabel="Porosity", legend=false)
p2 = plot(title = "Needles", 0:.1:1, bulks, ylabel="K/K0", xlabel="Porosity", legend=false)
plot(p1, p2)
```

```julia
nu0 = .45
nus = []
bulks = [] # K/K0
for porosity in 0:.1:.9
    props = sphere_properties(porosity, nu0)
    push!(nus, props[1])
    push!(bulks, props[2])
end 

p1 = plot(title = "Spheres", 0:.1:.9, nus, ylabel="nu", xlabel="Porosity", legend=false)
p2 = plot(title = "Spheres", 0:.1:.9, bulks, ylabel="K/K0", xlabel="Porosity", legend=false)
plot(p1, p2)
```

## Fluids

Implementing Han and Batzle's formulation 

```julia
# Should be able to go back and forth 

p = poissons(3000, 1000)
mu_from_poissons(1000, p)
```

```julia
# Example 
vp = 6
vpvs = 2
vs = vp/vpvs
rho = 3000 # kg/m^3
porosity=.05
Kf = 2.8 # GPa, for water
rho_f = 1000 # kg/m^3, for water
#Kf = 0 
#rho_f = 0

K0, mu0 = moduli_from_speeds(vp, vs, rho)
vpcheck, vscheck = speed_from_moduli(K0, mu0, rho)
println("check: $vp = $vpcheck; $vs = $vscheck")

println("K=$K0, mu=$mu0")
poissonsd, KdK0 = sphere_properties(porosity, poissons(K0, mu0))
Kd = KdK0*K0
mud = mu_from_poissons(Kd, poissonsd)
println("Kd = $Kd, mu_d = $mud")
vpd, vsd = speed_from_moduli(Kd, mud, (1-porosity)*rho)
println("dry speeds $vpd, $vsd")

Kw = fluid_sub(K0, Kf, Kd, porosity)
rho_w = wet_density(rho, rho_f, porosity)
println("K wet = $Kw, wet density $rho_w")

vpw, vsw = speed_from_moduli(Kw, mud, rho_w)
println("wet speeds $vpw, $vsw, vpvs = $(vpw/vsw)")
```

```julia
props = CrackProfile(cracked_properties, rho_f, Kf, .1, "w", "crack")
println(props)
apply_cracking(rho, vp, vpvs, props)
```
